```python
#Web Scraping Using Python

import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
# Creating Soup Object
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
#print(title)
```

```python
text = soup.get_text()
#print(soup.text)
```

```python
#soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
rows = soup.find_all('tr')
#print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
#removing the html tag using get_text method

str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
type(cleantext)
```

```python
#removing the html tag using regular expression
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
# Data manipulation and cleaning

df1 = df[0].str.split(',', expand=True)
df1.head(10)

```

```python
df1[0] = df1[0].str.strip('[')

df1.head(10)
```

```python
col_labels = soup.find_all('th')
#print(col_labels)
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
#converting list of header in pandas dataframe
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
#spliting it in multiple column

df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
#concating df1 and df3

frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
# assigning first row to table header

df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
df5.shape
```

```python
# droping rows with any missing value

df6 = df5.dropna(axis=0, how='any')
```

```python
#droping first index
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
#refining place and time
df7.rename(columns={'[Place': 'Place'},inplace=True)
df7.rename(columns={' Team]': 'Team'},inplace=True)
df7.head()
```

```python
#removing all "]" braces

df7['Team'] = df7['Team'].str.strip(']')
df7.head()
```
